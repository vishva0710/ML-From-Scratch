LINEAR REGRESSION

```python
#Linear Regression is one of the most basic and important Machine Learning algorithms used for predicting continuous numerical values.
#It finds the relationship between:
#Input variable (Feature)
#Output variable (Target)
#Example:
#Predicting house price
#Predicting salary
#Predicting student marks
#Predicting temperature
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
# let create data 
data =pd.DataFrame({'week':[1,2,3,4,5],'car_sold':[2,4,5,4,5]})
print(data)
```

```python
# the question is in the 6th week how many cars are sold ?
```

```python
# it is find the best fit line between the weeks and number of car_sold to find the relationship between them 
```

```python
# her we mark the point in the graph  
plt.scatter(data['week'],data['car_sold'])
plt.xlabel('week')
plt.ylabel('car_sold')
plt.show()
```

```python
# now we have to find the best fit line 
# the best fit line means - the line should touch or closer to the points 
# to find the best line we have to find the point of the line 
# the point is x=mean,y=mean 
```

```python
mean_x=np.mean(data['week'])
mean_y=np.mean(data['car_sold'])
print('mean of x',mean_x)
print('mean of y',mean_y)
```

```python
plt.scatter(data['week'],data['car_sold'])
plt.xlabel('week')
plt.ylabel('car_sold')
plt.axvline(x=mean_x,color='red',linestyle='--',label='mean_x')
plt.axhline(y=mean_y,color='red',linestyle='--',label='mean_y')
plt.legend()
plt.grid()
plt.show()
```

```python
# the intersect point of the mean_x and mean_Y  is a point of the best fit line 
# but we don't know the  direction of the line  
```

```python
# we have to find the direction if the line 
# for this we have to find the Σ(x - x̄)² (y - ȳ)²
```

```python
x=data['week']
y=data['car_sold']
summation=0
for i,j in zip(x,y):
  x_divation=mean_x-i
  y_divation=mean_y-j
  mul_x_y=x_divation*y_divation
  summation=mul_x_y+summation
print(summation)
```

```python
# now we have to find the line using line equation (y=mx+c)
#c=y-intercept,m=slop ,m= Σ(x - x̄)² (y - ȳ)²/Σ(x - x̄)²
```

```python
summation1=0
for k in x:
  x_divite=(mean_x-k)**2
  summation1=x_divite+summation1
print(summation1)
print(summation)  
#m= Σ(x - x̄)²(y - ȳ)²/Σ(x - x̄)²
m=summation/summation1
print(f'm-slop:{m}')
```

```python
# to find c value ȳ=𝑚x̄+𝑐
#ȳ=y_mean,x̄=x_mean
c=mean_y-(m*3)
print(c)
```

```python
#y=0.6x+2.2
#now we want to apply all the x value to find the predicted value y
value=[]
for x in data['week']:
  pred_y=(m*x)+c
  value.append(round(float(pred_y),1))
print(value)
```

```python
# now we get the preducted value 
plt.scatter(data['week'],data['car_sold'])
plt.xlabel('week')
plt.ylabel('car_sold')
plt.plot(data['week'],value,color='green',label='fit line')
plt.scatter(data['week'],value,color='red',marker='o',label='predicted point')
plt.axvline(x=mean_x,color='red',linestyle='--',label='mean_x')
plt.axhline(y=mean_y,color='red',linestyle='--',label='mean_y')
plt.legend()
plt.grid()
plt.show()
```

```python
#now we want to find the error by
#residual=actual point-predicted point
```

```python
actual_value=[]
predicted_Value=[]
for a,p in zip(data['car_sold'],value):
  actual_value.append(a)
  predicted_Value.append(p)
print(f'actual value:{actual_value}')
print(f' predicted value:{predicted_Value}')
```

```python
error=[]
for r2 in range(len(data['car_sold'])):
    residual=actual_value[r2]-predicted_Value[r2]
    error.append(round(residual,1))
print(error)
```

```python
square_error=[]
for e2 in range (len(error)):
    square_error.append(round(error[e2]*error[e2]))
print(square_error)
sum_of_error=sum(square_error)
print('error',sum_of_error)
```

```python
#totaly the error is  2 
# we can decrease the error by increasing  or decreasing the m and c value that can rotate the fit line towords the actual point  
```
