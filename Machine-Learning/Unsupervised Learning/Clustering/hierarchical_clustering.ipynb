```python
#do Standardization because all the features are in the different unites
from sklearn.preprocessing import StandardScaler
sds=StandardScaler()
x_stand=sds.fit_transform(x)
```

```python
x_stand
```

```python
from scipy.cluster.hierarchy import dendrogram,linkage
linkage=linkage(x_stand,method='ward')
```

```python
plt.Figure(figsize=(15,5))
dendrogram(linkage)
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
hc= AgglomerativeClustering(
  n_clusters=4,linkage='ward'
)
```

```python
from sklearn.metrics import silhouette_score

clusters = hc.fit_predict(x_stand)

score = silhouette_score(x_stand, clusters)

print(score)
```
