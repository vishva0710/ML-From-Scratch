```python
#Logistic Regression starts similarly to Linear Regression.
#It tries to learn the relationship between X (features) and Y (target).
#It uses weights (w) and bias (b).
#During training, it adjusts w and b using Gradient Descent to reduce the prediction error.
#After training, it calculates: z=wx+b

#Unlike Linear Regression, it does not use z as the final answer.

#Instead, it passes z through the Sigmoid Function: P=1/1+e^−z

#The sigmoid converts the output into a probability between 0 and 1.
#Then a threshold is applied (usually 0.5):
#Probability | Prediction
#-------------------------------
#≥ 0.5	     | Class 1 (Pass)
#< 0.5	     | Class 0 (Fail)
	
#Linear Regression predicts the final value directly.
#Logistic Regression first calculates z = wx+b.
#Then it converts z into a probability using the sigmoid function.
#The probability is used to decide the class.
#example :
#Hours Studied = 5


#z = wx + b
#z = 3

#Sigmoid(3)
#= 0.95

#0.95 > 0.5

#Prediction = Pass (1)
```

```python
import numpy as np
w=2
b=-7
x=5
#linear regression 
z=w*x+b
#sigmoid
probability = 1 / (1 + np.exp(-z))
# Classification
if probability>=0.5:
  print('pass')
else:
  print('fail')
print("Probability:", probability)
```
