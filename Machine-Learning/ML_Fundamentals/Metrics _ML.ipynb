```python
# in regression we have fore types 
# 1,--MAE (Mean Absolute Error) formula(MAE=n1​∑∣yi​−yi​^​)
# it is give the average error of the model 
# Example
# Actual	Predicted
#  10	       12
#  20	       18
#  30	       35

#Errors:

#|10 − 12| = 2
#|20 − 18| = 2
#|30 − 35| = 5

#MAE = (2 + 2 + 5) / 3 = 3

# Interpretation: On average, the model is making an error of 3 units.

#2.--MSE (Mean Squared Error)

#Formula: MSE=n1​∑(yi​−yi​^​)2

#Using the same errors:

#2² = 4
#2² = 4
#5² = 25

#MSE = (4 + 4 + 25)/3= 11

#Why square the errors:Because large errors get punished more.

#Error = 10 → 100
#Error = 20 → 400
#A model with huge mistakes gets a much worse score.
#3.,-- RMSE (Root Mean Squared Error)
#Formula: RMSE= root of MSE
	#RMSE = √11 ≈ 3.32

#Why use RMSE?

#MSE changes units because of squaring.

#For example:

#House price in lakhs
#MSE becomes lakh²

#RMSE converts it back to the original unit.

#4.-- R² Score (R-Squared)

#Measures how much variance the model explains.

#Range:

#1.0 → Perfect model
#0.9 → Excellent
#0.7 → Good
#0 → Same as guessing average
#Negative → Very poor model

#Example:

#R² = 0.85

#Means:

#The model explains 85% of the variation in the data.
```

```python
# Metrics for Classification
#When output is a category
#Examples:
#Spam / Not Spam
#Disease / No Disease
#Cat / Dog

#-- Accuracy

#Formula:Accuracy=Correct Predictions/Total Predictions
#Example:
# 100 samples
# 90 correct predictions

#Accuracy = 90 / 100= 90%

#Problem with Accuracy

#Suppose:

#990 healthy patients
# 10 sick patients

# Model predicts everyone is healthy.

#Accuracy:#990 / 1000 = 99%

#Looks great.

#But the model found 0 sick patients.

#So accuracy alone can be misleading


# CONFUSION MATRIX

# A Confusion Matrix is a table used to evaluate
# the performance of a classification model.

#                 Predicted
#               Yes      No
# Actual Yes     TP      FN
# Actual No      FP      TN

# TP (True Positive):
# Model correctly predicts Positive.

# TN (True Negative):
# Model correctly predicts Negative.

# FP (False Positive):
# Model predicts Positive, but actual is Negative.

# FN (False Negative):
# Model predicts Negative, but actual is Positive.


# EXAMPLE: DISEASE DETECTION

#                     Predicted
#                 Sick     Healthy
# Actual Sick       40        10
# Actual Healthy     5        45

# TP = 40
# FN = 10
# FP = 5
# TN = 45

# PRECISION

# Formula:
# Precision = TP / (TP + FP)

# Example:
# Precision = 40 / (40 + 5)
#           = 40 / 45
#           = 0.89
#           = 89%

# Meaning:
# When the model predicts Positive,
# it is correct 89% of the time.


# RECALL

# Formula:
# Recall = TP / (TP + FN)

# Example:
# Recall = 40 / (40 + 10)
#        = 40 / 50
#        = 0.80
#        = 80%

# Meaning:
# The model successfully found
# 80% of all actual Positive cases.


# F1 SCORE

# Formula:
# F1 Score = 2 * (Precision * Recall) / (Precision + Recall)

# F1 Score combines Precision and Recall
# into a single metric.

# Why use F1 Score?
# - Useful when classes are imbalanced.
# - Commonly used in:
#   * Fraud Detection
#   * Disease Detection
#   * Spam Email Detection
#
# A higher F1 Score indicates a better balance
# between Precision and Recall
```
