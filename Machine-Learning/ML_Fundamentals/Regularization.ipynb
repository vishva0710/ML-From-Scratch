Regularization

```python
# in machine Learning the data set  are divided into 4 parts x ,y,train,test x=feature,y=target,train= 80% of x and y,test=remaning 20% of data  
# bias = model makes too many assumptions and cannot learn the real pattern in the data.
# variance = model memorizes the training data instead of learning general patterns.
# Underfitting = high bias and low varance 
# Overfitting = low bias and high varance
# good fit =low bias and low varance 
```

```python
# Regularization is a technique used to reduce overfitting.
# there are three types of regularization L1(lasso),L2(ridge),L1 + L2 →(Elastic Net)  
```

```python
# overfit 
#| Experience | Salary |
#| 1          | 25     |
#| 2          | 30     |
#| 3          | 35     |
#| 4          | 40     |
#| 5          | 45     |

#Pattern: Salary = 20 + 5 × Experience

#| Experience | Salary |
#| 1          | 25     |
#| 2          | 30     |
#| 3          | 60     |
#| 4          | 40     |
#| 5          | 45     |

#Experience = 3 ,Salary = 60
#I must fit every point exactly
#So it bends and twists to pass through every training point

#Why Does Variance Become High?
# Because the model starts giving very large importance (weights) to tiny details.
#  Example:
#Salary = 5 × Experience
#Good.
#But overfitted:
#Salary =200 × Experience-150 × Age+300 × RandomFeature
#Weights become huge.
#The model becomes extremely sensitive.

#Training Data: Accuracy = 99%

#New Data: Accuracy = 60%

# The model memorized instead of learning.

#High Variance = Overfitting
```

```python
# Now Regularization Makes Sense
# Regularization says:
# Don't use huge weights.
#For example:
#Before:
#Experience = 200
#Age = -150
#RandomFeature = 300
#After regularization:
#Experience = 5
#Age = 0.2
#RandomFeature = 0
#The model becomes simpler and focuses on real patterns.
```

```python
#L1 (Lasso)
#Pushes some weights all the way to: 0

#Example:

#Experience = 5
#Age = 0
#RandomFeature = 0

# Removes unimportant features.

#How does Lasso know that a feature is unwanted and should get a weight of 0?"

#Let's use an example

#Suppose you're predicting salary.

#Experience	Favorite Color	Salary
#1	           Red	          25
#2	           Blue	          30
#3	           Green	        35
#4	           Red	          40
#5	           Blue	          45

#Notice:

#As Experience increases, Salary increases. 
#Favorite Color has no consistent relationship with Salary. 
#During Training

#The model tries different weights.

#For example:

#Salary =5 × Experience+ 10 × Favorite_Color

#The model checks:

#"Does Favorite Color help me predict Salary better?"

#The answer is No.

#So the weight on Favorite Color isn't helping reduce prediction error.
```

```python
#L2 (Ridge)

#Makes weights smaller but usually not exactly zero.

#Example:

#Experience = 5
#Age = 0.5
#RandomFeature = 0.1

#Keeps features but reduces their influence.
```

```python
#Data
#  ↓
#Model becomes too complex
#  ↓
#Large weights
#  ↓
#High Variance (Overfitting)
#  ↓
#Poor performance on new data
#  ↓
#Use Regularization
#  ↓
#Reduce large weights
#  ↓
#Better Generalization
```
